```python
import numpy as np
```

```python
from PIL import Image, ImageFont, ImageDraw, ImageEnhance

source_img = Image.open('Temp_quant.jpg')
draw = ImageDraw.Draw(source_img)
```

```python
from tqdm import tqdm
```

```python
bounding_box = [300, 320, 800, 385]
x1, y1, x2, y2 = bounding_box  # For easy reading

font = ImageFont.truetype('PTF55F.ttf', size=30)

names_list = np.load('swd_names.npy')
```

```python
import os
```

```python
try:
    os.mkdir('certs_quant')
except FileExistsError:
    pass
```

```python
for elem in tqdm(names_list):
    source_img = Image.open('Temp_quant.jpg')
    draw = ImageDraw.Draw(source_img)
    message = elem[1]
    w, h = draw.textsize(message, font=font)
    x = (x2 - x1 - w)/2 + x1
    y = (y2 - y1 - h)/2 + y1
    draw.text((x, y), message, align = 'center', fill = 'red', font = font)
    source_img.save('certs_quant/{}.jpg'.format(elem[0]))
```
